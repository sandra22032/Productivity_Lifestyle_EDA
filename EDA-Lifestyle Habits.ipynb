# VISUALIZATION

```python
# Setup: imports, theme, ordinal orders

import warnings
warnings.filterwarnings("ignore")

sns.set_theme(style="whitegrid", palette="muted", font_scale=1.15)

# Ordinal category orders (used across charts)
sleep_order    = ["Less than 5 hours","5-6 hours","6-7 hours","7-8 hours","More than 8 hours"]
screen_order   = ["Less than 1 hour","1-2 hours","3-4 hours","5-6 hours","More than 6 hours"]
exercise_order = ["Never","Rarely","1-2 times a week","3-5 times a week","Daily"]
phone_order    = ["Never","Rarely","Sometimes","Often","Always"]
plan_order     = ["Never","Rarely","Occasionally","Several times a week","Daily"]
age_order      = ["Below 18","18-20","21-23","24-26","27-29","30+"]

# Numeric encoding for correlation chart
data["sleep_num"]      = pd.Categorical(data["sleep_hours"], categories=sleep_order, ordered=True).codes
data["screen_num"]     = pd.Categorical(data["daily_screen_time"], categories=screen_order, ordered=True).codes
data["exercise_num"]   = pd.Categorical(data["exercise_frequency"], categories=exercise_order, ordered=True).codes
data["phone_num"]      = pd.Categorical(data["phone_check_frequency_morning"], categories=phone_order, ordered=True).codes
data["plan_num"]       = pd.Categorical(data["daily_planning_frequency"], categories=plan_order, ordered=True).codes
data["age_group"]      = pd.Categorical(data["age_group"], categories=age_order, ordered=True)

print("✓ Setup complete — data shape:", data.shape)
```

```python
# ── CHART 1: Univariate Overview (3×3 Dashboard) ──────────────────────────
# Shows distributions of all key variables at a glance

fig, axes = plt.subplots(3, 3, figsize=(25, 18))
fig.suptitle("Productivity & Lifestyle Habits — Overview", fontsize=20, fontweight="bold", y=1.01)

PALETTE = "Set2"

#------------------------------------------------------------------------------------------------------------------------------

# Age Group
ax = axes[0, 0]
counts = data["age_group"].value_counts().sort_index()
bars = sns.barplot(x=counts.index.astype(str), y=counts.values, ax=ax, palette=PALETTE, edgecolor="white")
ax.set_title("Age Group Distribution", fontweight="bold")
ax.set_xlabel("Age Group"); ax.set_ylabel("Count")
for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            int(bar.get_height()), ha="center", va="bottom", fontsize=10, fontweight="bold")

#------------------------------------------------------------------------------------------------------------------------------

# Gender Split
ax = axes[0, 1]
vc = data["gender"].value_counts()
ax.pie(vc.values, labels=vc.index, autopct="%1.1f%%",
       colors=["#92C5DE", "#F4A582"], startangle=90,
       wedgeprops={"edgecolor": "white", "linewidth": 2})
ax.set_title("Gender Split", fontweight="bold")

#------------------------------------------------------------------------------------------------------------------------------

# Productivity Rating Histogram
ax = axes[0, 2]
sns.histplot(data["productivity_rating"], bins=10, kde=True, ax=ax,
             color="#4C72B0", edgecolor="white", alpha=0.8)
ax.axvline(data["productivity_rating"].mean(), color="crimson",
           linestyle="--", linewidth=2, label=f"Mean = {data['productivity_rating'].mean():.1f}")
ax.set_title("Productivity Rating", fontweight="bold")
ax.set_xlabel("Rating (1–10)"); ax.legend(fontsize=10)

#------------------------------------------------------------------------------------------------------------------------------

# Sleep Hours
ax = axes[1, 0]
s_counts = data["sleep_hours"].value_counts().reindex(sleep_order)
sns.barplot(x=s_counts.values, y=s_counts.index, ax=ax, palette="Blues_d",
            orient="h", edgecolor="white")
ax.set_title("Sleep Hours", fontweight="bold"); ax.set_xlabel("Count")

#------------------------------------------------------------------------------------------------------------------------------

# Exercise Frequency
ax = axes[1, 1]
e_counts = data["exercise_frequency"].value_counts().reindex(exercise_order)
sns.barplot(x=e_counts.values, y=e_counts.index, ax=ax, palette="Greens_d",
            orient="h", edgecolor="white")
ax.set_title("Exercise Frequency", fontweight="bold"); ax.set_xlabel("Count")

#------------------------------------------------------------------------------------------------------------------------------

# Daily Screen Time
ax = axes[1, 2]
sc_counts = data["daily_screen_time"].value_counts().reindex(screen_order)
sns.barplot(x=sc_counts.values, y=sc_counts.index, ax=ax, palette="Oranges_d",
            orient="h", edgecolor="white")
ax.set_title("Daily Screen Time", fontweight="bold"); ax.set_xlabel("Count")

#------------------------------------------------------------------------------------------------------------------------------

# Productivity Challenges
ax = axes[2, 0]
ch = data["productivity_challenge"].value_counts().head(8)
sns.barplot(y=ch.index, x=ch.values, ax=ax, palette=PALETTE, orient="h", edgecolor="white")
ax.set_title("Top Productivity Challenges", fontweight="bold"); ax.set_xlabel("Count")

#------------------------------------------------------------------------------------------------------------------------------

# Preferred Productivity Tool
ax = axes[2, 1]
pt = data["preferred_productivity_tool"].value_counts()
wedges, _, autotexts = ax.pie(pt.values, labels=None, autopct="%1.0f%%",
    colors=sns.color_palette(PALETTE, len(pt)), startangle=90,
    wedgeprops={"edgecolor": "white", "linewidth": 1.5})
ax.legend(wedges, pt.index, loc="upper center", bbox_to_anchor=(0.5, -0.05), fontsize=8, ncol=2)
ax.set_title("Preferred Productivity Tool", fontweight="bold")

#------------------------------------------------------------------------------------------------------------------------------

# Most Used Platform (Top 8)
ax = axes[2, 2]
mp = data["most_used_platform"].value_counts().head(8)
sns.barplot(y=mp.index, x=mp.values, ax=ax, palette=PALETTE, orient="h", edgecolor="white")
ax.set_title("Most Used Platform (Top 8)", fontweight="bold"); ax.set_xlabel("Count")

#------------------------------------------------------------------------------------------------------------------------------

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
# ── CHART 2: Correlation Heatmap (Ordinal-Encoded Variables) ──────────────
# Reveals which lifestyle habits correlate most with productivity

num_cols = ["productivity_rating", "sleep_num", "screen_num",
            "exercise_num", "phone_num", "plan_num"]
labels   = ["Productivity\nRating", "Sleep\nHours", "Screen\nTime",
            "Exercise\nFreq", "Phone Check\n(Morning)", "Planning\nFreq"]

corr = data[num_cols].corr()
corr.index = labels
corr.columns = labels

fig, ax = plt.subplots(figsize=(9, 7))

sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
            vmin=-1, vmax=1, ax=ax, linewidths=0.8,
            linecolor="white", annot_kws={"size": 12, "weight": "bold"},
            cbar_kws={"shrink": 1, "label": "Pearson r"})

ax.set_title("Correlation Matrix — Lifestyle & Productivity",
             fontsize=15, fontweight="bold", pad=15)

plt.tight_layout()
plt.show()
```

```python
# ── CHART 3: Bivariate — Productivity Rating vs Key Lifestyle Variables ────
# Box / violin / bar plots showing how each habit links to productivity score

fig, axes = plt.subplots(2, 3, figsize=(20, 11))
fig.suptitle("Productivity Rating vs Lifestyle Factors",
             fontsize=18, fontweight="bold", y=1.01)

#------------------------------------------------------------------------------------------------------------------------------

# Sleep → Productivity
ax = axes[0, 0]
sns.boxplot(data=data, x="sleep_hours", y="productivity_rating",
            order=sleep_order, palette="Blues", ax=ax,
            linewidth=1.5, flierprops={"marker": "o", "markersize": 5})
ax.set_title("Sleep Hours → Productivity", fontweight="bold")
ax.set_xlabel("Sleep Hours"); ax.set_ylabel("Productivity Rating")
ax.tick_params(axis='x', rotation=20)

#------------------------------------------------------------------------------------------------------------------------------

# Exercise → Productivity
ax = axes[0, 1]
sns.violinplot(data=data, x="exercise_frequency", y="productivity_rating",
               order=exercise_order, palette="Greens", ax=ax, inner="box")
ax.set_title("Exercise Frequency → Productivity", fontweight="bold")
ax.set_xlabel("Exercise Frequency"); ax.set_ylabel("Productivity Rating")
ax.tick_params(axis='x', rotation=20)

#------------------------------------------------------------------------------------------------------------------------------

# Screen Time → Productivity
ax = axes[0, 2]
sns.boxplot(data=data, x="daily_screen_time", y="productivity_rating",
            order=screen_order, palette="Oranges", ax=ax,
            linewidth=1.5, flierprops={"marker": "o", "markersize": 5})
ax.set_title("Screen Time → Productivity", fontweight="bold")
ax.set_xlabel("Daily Screen Time"); ax.set_ylabel("Productivity Rating")
ax.tick_params(axis='x', rotation=20)

#------------------------------------------------------------------------------------------------------------------------------

# Current Status → Avg Productivity
ax = axes[1, 0]
status_order = (data.groupby("current_status")["productivity_rating"]
                    .mean().sort_values(ascending=False).index)
sns.barplot(data=data, x="current_status", y="productivity_rating",
            order=status_order, palette="Set2", ax=ax,
            capsize=0.1, err_kws={"linewidth": 1.5}, edgecolor="white")
ax.set_title("Current Status → Avg Productivity", fontweight="bold")
ax.set_xlabel(""); ax.set_ylabel("Mean Productivity Rating")
ax.tick_params(axis='x', rotation=25)

#------------------------------------------------------------------------------------------------------------------------------

# Planning Frequency → Productivity
ax = axes[1, 1]
sns.barplot(data=data, x="daily_planning_frequency", y="productivity_rating",
            order=plan_order, palette="Purples_d", ax=ax,
            capsize=0.1, err_kws={"linewidth": 1.5}, edgecolor="white")
ax.set_title("Planning Frequency → Productivity", fontweight="bold")
ax.set_xlabel("Planning Frequency"); ax.set_ylabel("Mean Productivity Rating")
ax.tick_params(axis='x', rotation=20)

#------------------------------------------------------------------------------------------------------------------------------

# Productivity Tool → Avg Productivity
ax = axes[1, 2]
tool_order = (data.groupby("preferred_productivity_tool")["productivity_rating"]
                  .mean().sort_values(ascending=False).index)
sns.barplot(data=data, x="preferred_productivity_tool", y="productivity_rating",
            order=tool_order, palette="Set2", ax=ax,
            capsize=0.1, err_kws={"linewidth": 1.5}, edgecolor="white")
ax.set_title("Productivity Tool → Avg Rating", fontweight="bold")
ax.set_xlabel("")
ax.tick_params(axis='x', rotation=25)
for label in ax.get_xticklabels():
    label.set_ha("right")

#------------------------------------------------------------------------------------------------------------------------------

plt.tight_layout(h_pad=5, w_pad=5)
plt.show()
```

```python
# ── CHART 4: Gender & Age Group Comparisons ───────────────────────────────
# How demographics shape productivity, sleep, and exercise behaviour

fig, axes = plt.subplots(2, 2, figsize=(16, 11))
fig.suptitle("Gender & Age Group Comparisons", fontsize=18, fontweight="bold")

# ── 4a. Gender vs Productivity (box) ──
ax = axes[0, 0]
sns.boxplot(data=data, x="gender", y="productivity_rating",
            palette=["#F4A582", "#92C5DE"], ax=ax, width=0.5,
            linewidth=1.8, flierprops={"marker": "o", "markersize": 5})
ax.set_title("Gender vs Productivity Rating", fontweight="bold")
ax.set_xlabel("Gender"); ax.set_ylabel("Productivity Rating")

# ── 4b. Age Group vs Productivity (box) ──
ax = axes[0, 1]
sns.boxplot(data=data, x="age_group", y="productivity_rating",
            order=age_order, palette="Set2", ax=ax,
            linewidth=1.5, flierprops={"marker": "o", "markersize": 5})
ax.set_title("Age Group vs Productivity Rating", fontweight="bold")
ax.set_xlabel("Age Group"); ax.set_ylabel("Productivity Rating")

# ── 4c. Exercise Frequency by Gender (grouped % bar) ──
ax = axes[1, 0]
cross = (data.groupby(["gender", "exercise_frequency"])
             .size().reset_index(name="count"))
cross["pct"] = cross.groupby("gender")["count"].transform(lambda x: x / x.sum() * 100)
pivot = cross.pivot(index="exercise_frequency", columns="gender", values="pct").reindex(exercise_order)
pivot.plot(kind="bar", ax=ax, color=["#F4A582", "#92C5DE"],
           width=0.7, edgecolor="white")
ax.set_title("Exercise Frequency by Gender (%)", fontweight="bold")
ax.set_xlabel("Exercise Frequency"); ax.set_ylabel("% Within Gender")
ax.legend(title="Gender"); ax.tick_params(axis='x', rotation=25)

# ── 4d. Sleep Hours by Age Group (stacked 100% bar) ──
ax = axes[1, 1]
sleep_age = (data.groupby(["age_group", "sleep_hours"])
                 .size().reset_index(name="count"))
sleep_age["pct"] = sleep_age.groupby("age_group")["count"] \
                             .transform(lambda x: x / x.sum() * 100)
pivot2 = (sleep_age.pivot(index="age_group", columns="sleep_hours", values="pct")
                   .reindex(columns=sleep_order).fillna(0))
pivot2.plot(kind="bar", stacked=True, ax=ax,
            color=sns.color_palette("Blues_d", len(sleep_order)),
            edgecolor="white", width=0.75)
ax.set_title("Sleep Hours by Age Group (% stacked)", fontweight="bold")
ax.set_xlabel("Age Group"); ax.set_ylabel("Percentage")
ax.legend(title="Sleep Hours", bbox_to_anchor=(1, 1), fontsize=8)
ax.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
# ── CHART 5: Screen Time Deep Dive ────────────────────────────────────────
# How screen time links to productivity impact perception & platform choice

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle("Screen Time Deep Dive", fontsize=18, fontweight="bold")

# ── 5a. Does screen time affect productivity? (stacked % bar by screen band) ──
ax = axes[0]
cross2 = (data.groupby(["daily_screen_time", "screen_time_affects_productivity"])
              .size().reset_index(name="count"))
cross2["pct"] = cross2.groupby("daily_screen_time")["count"] \
                       .transform(lambda x: x / x.sum() * 100)
pivot3 = (cross2.pivot(index="daily_screen_time",
                       columns="screen_time_affects_productivity", values="pct")
                .reindex(screen_order)[["No", "Not sure", "Yes"]].fillna(0))
pivot3.plot(kind="bar", stacked=True, ax=ax,
            color=["#2ca02c", "#f0c040", "#d62728"], edgecolor="white", width=0.75)
ax.set_title("Does Screen Time Affect Productivity?\n(by Screen Time Band)", fontweight="bold")
ax.set_xlabel("Daily Screen Time"); ax.set_ylabel("% of Band")
ax.tick_params(axis='x', rotation=25)
ax.legend(title="Affects Prod?")

# ── 5b. Most Used Platform (count bar) ──
ax = axes[1]
mp = data["most_used_platform"].value_counts().head(8)
colors = sns.color_palette("Set2", len(mp))
bars = ax.barh(mp.index, mp.values, color=colors, edgecolor="white")
ax.set_title("Most Used Platform (Top 8)", fontweight="bold")
ax.set_xlabel("Count")
for bar, val in zip(bars, mp.values):
    ax.text(val + 0.2, bar.get_y() + bar.get_height()/2,
            str(val), va="center", fontsize=10, fontweight="bold")

# ── 5c. Avg Productivity by Platform (top 8) ──
ax = axes[2]
top_platforms = data["most_used_platform"].value_counts().head(8).index
plat_df  = data[data["most_used_platform"].isin(top_platforms)]
plat_prod = plat_df.groupby("most_used_platform")["productivity_rating"] \
                    .mean().sort_values(ascending=False)
bars2 = ax.barh(plat_prod.index, plat_prod.values,
                color=sns.color_palette("muted", len(plat_prod)), edgecolor="white")
ax.axvline(data["productivity_rating"].mean(), color="crimson",
           linestyle="--", linewidth=1.8, label=f"Overall Mean ({data['productivity_rating'].mean():.1f})")
ax.set_title("Avg Productivity by Platform (Top 8)", fontweight="bold")
ax.set_xlabel("Mean Productivity Rating")
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# ── CHART 6: Productivity Challenges & Influencing Factors ────────────────
# Crosstab heatmap + grouped bar to see who faces what challenges

fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle("Productivity Challenges & Influencing Factors",
             fontsize=18, fontweight="bold")

# ── 6a. Challenge × Current Status (heatmap of counts) ──
ax = axes[0]
top_ch   = data["productivity_challenge"].value_counts().head(8).index
ch_cross = pd.crosstab(data["productivity_challenge"],
                        data["current_status"]).loc[top_ch]
sns.heatmap(ch_cross, annot=True, fmt="d", cmap="YlOrRd",
            linewidths=0.6, linecolor="white", ax=ax,
            annot_kws={"size": 11, "weight": "bold"},
            cbar_kws={"shrink": 0.8})
ax.set_title("Challenges × Current Status (Count)", fontweight="bold")
ax.set_xlabel("Current Status"); ax.set_ylabel("Challenge")
ax.tick_params(axis='x', rotation=25)

# ── 6b. Productivity Factor × Gender (grouped bar) ──
ax = axes[1]
factor_gender = (data.groupby(["productivity_factor", "gender"])
                     .size().reset_index(name="count"))
sns.barplot(data=factor_gender, x="productivity_factor", y="count",
            hue="gender", palette=["#F4A582", "#92C5DE"],
            ax=ax, edgecolor="white")
ax.set_title("Key Productivity Factors by Gender", fontweight="bold")
ax.set_xlabel("Productivity Factor"); ax.set_ylabel("Count")
ax.tick_params(axis='x', rotation=25)
ax.legend(title="Gender")

plt.tight_layout()
plt.show()
```

```python
# ── CHART 7: Morning Phone Check Behaviour ────────────────────────────────
# Links early-morning phone usage to productivity and planning habits

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle("Morning Phone Check Behaviour", fontsize=18, fontweight="bold")

# ── 7a. Phone Check Frequency → Productivity (box) ──
ax = axes[0]
sns.boxplot(data=data, x="phone_check_frequency_morning",
            y="productivity_rating", order=phone_order,
            palette="Reds", ax=ax, linewidth=1.5,
            flierprops={"marker": "o", "markersize": 5})
ax.set_title("Morning Phone Check → Productivity Rating", fontweight="bold")
ax.set_xlabel("Check Frequency"); ax.set_ylabel("Productivity Rating")

# Add mean line for reference
mean_val = data["productivity_rating"].mean()
ax.axhline(mean_val, color="navy", linestyle="--",
           linewidth=1.5, label=f"Overall Mean ({mean_val:.1f})")
ax.legend(fontsize=9)

# ── 7b. Phone Check × Planning Frequency (stacked 100% bar) ──
ax = axes[1]
ph_cross = (data.groupby(["phone_check_frequency_morning", "daily_planning_frequency"])
                .size().reset_index(name="count"))
ph_cross["pct"] = ph_cross.groupby("phone_check_frequency_morning")["count"] \
                           .transform(lambda x: x / x.sum() * 100)
ph_pivot = (ph_cross.pivot(index="phone_check_frequency_morning",
                            columns="daily_planning_frequency", values="pct")
                    .reindex(phone_order)[plan_order].fillna(0))
ph_pivot.plot(kind="bar", stacked=True, ax=ax,
              color=sns.color_palette("Purples_d", 5), edgecolor="white", width=0.75)
ax.set_title("Morning Phone Check × Daily Planning Frequency (%)", fontweight="bold")
ax.set_xlabel("Check Frequency"); ax.set_ylabel("Percentage")
ax.tick_params(axis='x', rotation=20)
ax.legend(title="Plan Freq", bbox_to_anchor=(1, 1), fontsize=8)

plt.tight_layout()
plt.show()
```

# Key Insights to Highlight

1. Daily exercise is the single strongest positive lifestyle signal — Daily exercisers average 7.44/10 vs 4.00/10 for those who never exercise. A clear monotonic gradient.
2. Screen time inversely tracks productivity — Less than 1 hr/day → mean 7.80; 6+ hrs/day → mean 5.39. And 90%+ of high-screen-time users acknowledge it hurts them.
3. Planning discipline matters more than sleep — Daily planners average 7.31 vs 4.43 for "Rarely" planners. Sleep correlation is surprisingly near-zero (–0.014), suggesting sleep quality or timing (not just hours) may matter more than raw duration.
4. Morning phone checking is a red flag — "Always" checkers cluster at lower productivity scores and are also the least likely to plan their day — suggesting a compulsive vs intentional morning routine divide.
5. Procrastination is a student problem; stress is a professional problem — The challenge heatmap clearly separates these two groups.
6. Instagram is dominant but not uniquely harmful — It's the most used platform by far, but productivity averages across platforms are similar, suggesting the platform itself matters less than total usage time.
7. Tool users outperform non-users — Calendar/Planner users average higher than those who use no productivity tools at all, reinforcing the planning frequency finding.
